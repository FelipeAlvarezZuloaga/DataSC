# Bases de datos en Python

Es clave saber conectarse a bases de datos desde diferentes softwares. Sin embargo, la mayoría de las veces vamos a requerir procesar los datos de la base y para eso requerimos traer los datos a Python.

Vamos a hacer la misma conexión que hicimos con DBeaver pero ahora directamente en Python

```python
from sqlalchemy import create_engine, text #Tambien requerimos **pip install pymysql
import pandas as pd
```

Levantamos la conexión al paserle los valores de configuración de la base de datos

```python
db_config = {
        
            "username" : "anonymous",
            "password" : "",
            "host" : "ensembldb.ensembl.org",
            "port" : "3306",
            "db" : "homo_sapiens_core_109_38",
            "table" : "gene",
}
```

```python
# Formato de la cadena de conexión:
# 'mysql+pymysql://usuario:contraseña@host:puerto/nombre_base_datos'

engine = create_engine(f'mysql+pymysql://{db_config["username"]}:{db_config["password"]}@{db_config["host"]}:{db_config["port"]}/{db_config["db"]}')
```

```python
# Hagámos un querry ara bajar TODA la información de la tabla

query = text(f"SELECT * FROM {db_config["table"]}")
```

```python
# Corremos el query y lo guardamos en una variable con pandas

data = pd.read_sql(query, engine)
```
